```python
# subset data for cells that contain 'MARIJUANA' within the 'Descript' column
# and save it to a dataframe called historical_data_marijuana
historical_data_marijuana = historical_data[
    historical_data['Descript'].str.contains('MARIJUANA')
].copy()

# df[
#     # expression
# ]
```
